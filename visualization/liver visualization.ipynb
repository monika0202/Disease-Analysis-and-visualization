```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import pandas as pd
import seaborn as sns
```

```python
data=pd.read_csv('/home/dell/python_machine_learning/disease/csv/liver.csv')
data.head()
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(25,10))
data['Age'].value_counts().plot.bar(color='darkviolet')
```

```python
plt.rcParams['figure.figsize']=(10,10)
sns.pairplot(data,hue='Gender')
```

```python
sns.pairplot(data)
```

```python
ax = plt.subplots(figsize=(8, 6))
sns.scatterplot(x="Albumin", y="Albumin_and_Globulin_Ratio",color='mediumspringgreen',data=data);
plt.show()
```

```python
plt.figure(figsize=(8,6))
data.groupby('Gender').sum()["Total_Protiens"].plot.bar(color='coral')
```

```python
corr=data.corr()

plt.figure(figsize=(20,10)) 
sns.heatmap(corr,cmap="Greens",annot=True)
```

```python
xx=data['Age'].values.reshape(1,-1)
yy=data['Aspartate_Aminotransferase'].values.reshape(1,-1)
zz=data['Gender'].values.reshape(1,-1)
```

```python
sns.set_style("white")
fig = plt.figure(figsize=(12,9))
ax = plt.axes(projection="3d")

ax.scatter3D(xx, yy, data['Alkaline_Phosphotase'],cmap='hsv',s=200, c="#1aff8c")




ax.set_xlabel("Age")
ax.set_ylabel("Aspartate_Aminotransferase")
ax.set_zlabel("Alkaline_Phosphotase")

#ax.plot_surface(xx, yy, zz, alpha=0.4)
plt.show()
```

```python
labels = data.Gender.value_counts().index
colors = ['green','yellow']
explode = [0,0]
sizes = data.Gender.value_counts().values

plt.figure(0,figsize = (7,7))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('Percentage of Male and Female',fontsize = 15)
```
